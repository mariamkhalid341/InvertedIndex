```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.action_chains import ActionChains
import time
import logging
import pdb
import shutil
import os
```

```python
landing_folder = "D:\Coventry"

prefrences = {"download.default_directory": landing_folder}
```

```python
class Crawler:
    
    def __init__(self):
        
        self.oxford_ds_professors = ['https://www.oii.ox.ac.uk/people/mark-graham/?profile',
                               'https://www.oii.ox.ac.uk/people/bernie-hogan/?profile',
                              'https://www.oii.ox.ac.uk/people/brent-mittelstadt/?profile',
                              'https://www.oii.ox.ac.uk/people/gina-neff/?profile',
                              'https://www.oii.ox.ac.uk/people/ralph-schroeder/?profile',
                           'https://www.oii.ox.ac.uk/people/mariarosaria-taddeo/?profile',
                           'https://www.oii.ox.ac.uk/people/sandra-wachter/?profile']
        
        self.coventry_ds_professors = ['https://pureportal.coventry.ac.uk/en/persons/vasile-palade',
                                      'https://pureportal.coventry.ac.uk/en/persons/alireza-daneshkhah',
                                      'https://pureportal.coventry.ac.uk/en/persons/nader-sohrabi-safa',
                                      'https://pureportal.coventry.ac.uk/en/persons/james-brusey',
                                      'https://pureportal.coventry.ac.uk/en/persons/ye-liu',
                                      'https://pureportal.coventry.ac.uk/en/persons/elena-gaura']
        
        self.kingston_arts_professors = ['https://www.kingston.ac.uk/staff/profile/dr-hager-weslati-662/',
                                        'https://www.kingston.ac.uk/staff/profile/professor-john-oacute-maoilearca-124/',
                                        'https://www.kingston.ac.uk/staff/profile/dr-matthew-melia-648/',
                                        'https://www.kingston.ac.uk/staff/profile/dr-mariacutea-menciacutea-629/',
                                        'https://www.kingston.ac.uk/staff/profile/dr-landeacute-pratt-598/']
        
        #'https://www.kingston.ac.uk/staff/profile/professor-alex-hill-71/',
        self.kingston_business_professors = ['https://www.kingston.ac.uk/staff/profile/dr-reza-zanjirani-farahani-341/',
                                        'https://www.kingston.ac.uk/staff/profile/dr-george-alexandrou-432/',
                                        'https://www.kingston.ac.uk/staff/profile/dr-rahul-chawdhary-146/',
                                        'https://www.kingston.ac.uk/staff/profile/dr-john-sebastian-pereira-356/',
                                        'https://www.kingston.ac.uk/staff/profile/dr-pauline-parker-932/',
                                        'https://www.kingston.ac.uk/staff/profile/dr-fatima-annan-diab-381/',
                                        'https://www.kingston.ac.uk/staff/profile/dr-barry-avery-147/',
                                        'https://www.kingston.ac.uk/staff/profile/dr-marvyn-boatswain-126/']
        
    
    def setupChrome(self):
        
        self.settings = webdriver.ChromeOptions()
        self.settings.add_argument("--incognito")
        self.settings.add_experimental_option('prefs',prefrences)
    
    def loadBrowser(self):
        
        self.setupChrome()

        try:
            self.browser = webdriver.Chrome(chrome_options=self.settings, executable_path=r"D:\Coventry\chromedriver.exe")
            self.browser.maximize_window()

        except Exception as e:
            self.logger.critical("Unable to load chrome driver. " + str(e))
    
    def setupPage(self):

        self.setupChrome()
        self.loadBrowser()
        
    def crawl_oxford_professors_information(self, professor_link):
        # extracting each professor information
        self.browser.get(professor_link)
        
        #print("in crawler")
        profile_text = "//*[@id='profile']"
        page_text = self.browser.find_element_by_xpath(profile_text).text
        
        # profile text
        #print(page_text)
        professor_name = professor_link[professor_link.index('people/') + 7 : professor_link.index('/?profile')]
        print(professor_name)
        prof_file = open(professor_name + ".txt", "a+")
        
        prof_file.write(page_text)
        prof_file.close()
        
        #Research Text
        time.sleep(1)
        research_button = "//*[@id='main-content']/div[2]/div/div/div[2]/ul/button[2]"
        self.browser.find_element_by_xpath(research_button).click()
        
        research = "//*[@id='main-content']/div[2]/div/div/div[2]"
        time.sleep(2)
        research_text = self.browser.find_element_by_xpath(research).text
        
        prof_file = open(professor_name + ".txt", "a+")
        prof_file.write(" " + research_text)
        prof_file.close()
        
    def crawl_coventry_professors_information(self, professor_link):
        # extracting each professor information
        self.browser.get(professor_link)
        
        print("in crawler")
        profile_text = "//*[@id='main-content']/section[1]/div/div/div[1]/div"
        page_text = self.browser.find_element_by_xpath(profile_text).text
        
        # profile text
        #print(page_text)
        professor_name = professor_link[professor_link.index('persons/') +8 :]
        print(professor_name)
        prof_file = open(professor_name + ".txt", "a+")
        
        prof_file.write(page_text)
        prof_file.close()
        
        #Research Text

    def crawl_KingstonArts_professors_information(self, professor_link):
        # extracting each professor information
        self.browser.get(professor_link)
        
        print("in crawler")
        profile_text = "//*[@id='about-jump']/div/div[1]"
        page_text = self.browser.find_element_by_xpath(profile_text).text
        
        # profile text
        #print(page_text)
        professor_name = professor_link[professor_link.index('profile/') +8 :][:-5]
        print(professor_name)
        prof_file = open(professor_name + ".txt", "a+")
        
        prof_file.write(page_text)
        prof_file.close()
        
        #Research Text
        time.sleep(2)
        research_button = "//*[@id='domain-research-jump']/div/h3"
        self.browser.find_element_by_xpath(research_button).click()
        
        research = "//*[@id='middle-col']/div[6]/div/div/div/div[4]/div"
        time.sleep(2)
        research_text = self.browser.find_element_by_xpath(research).text
        
        prof_file = open(professor_name + ".txt", "a+")
        prof_file.write(" " + research_text)
        prof_file.close()
        
    def crawl_KingstonBusiness_professors_information(self, professor_link):
        # extracting each professor information
        self.browser.get(professor_link)
        
        print("in crawler")
        profile_text = "//*[@id='about-jump']/div/div[1]"
        page_text = self.browser.find_element_by_xpath(profile_text).text
        
        # profile text
        #print(page_text)
        professor_name = professor_link[professor_link.index('profile/') +8 :][:-1]
        print(professor_name)
        prof_file = open(professor_name + ".txt", "a+")
        
        prof_file.write(page_text)
        prof_file.close()
        
        #Research Text
        time.sleep(2)
        research_button = "//*[@id='domain-research-jump']/div/h3"
        self.browser.find_element_by_xpath(research_button).click()
        time.sleep(1)
        body = self.browser.find_element_by_css_selector('body')
        body.send_keys(Keys.PAGE_DOWN)
        research = "//*[@id='middle-col']/div[6]/div/div/div/div[4]/div"
        time.sleep(2)
        research_text = self.browser.find_element_by_xpath(research).text
        
        prof_file = open(professor_name + ".txt", "a+")
        prof_file.write(" " + research_text)
        prof_file.close()
        
    def iterateOverInputs(self,uni):
        
        if uni == 'kingston':
            for i in self.kingston_arts_professors:

                self.professor_link = i

                time.sleep(1)
                self.crawl_KingstonArts_professors_information(self.professor_link)
        
        if uni == 'oxford':
            for i in self.oxford_ds_professors:

                self.professor_link = i

                time.sleep(1)
                self.crawl_oxford_professors_information(self.professor_link)
            
        if uni == 'coventry':
            for i in self.coventry_ds_professors:

                self.professor_link = i

                time.sleep(1)
                self.crawl_coventry_professors_information(self.professor_link)     
                
        if uni == 'kingstonBusiness':
            for i in self.kingston_business_professors:

                self.professor_link = i

                time.sleep(1)
                self.crawl_KingstonBusiness_professors_information(self.professor_link)            
```

```python
# main function 

if __name__ == '__main__':
    obj = Crawler()
    obj.setupPage()
    obj.iterateOverInputs('kingstonBusiness')
```
